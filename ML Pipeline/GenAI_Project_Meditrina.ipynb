```python
## Text to Speech conversion

def readn(nstr):
    engine = pyttsx3.init()   # Initialize the text-to-speech engine

    engine.setProperty('voice', "english+f5")   # Set the voice property (specific voice identifier)
    engine.setProperty('rate', 130)   # Set the speech rate property (words per minute) :: 130 is slow for clear audio

    engine.say(nstr)   # Queue the text to be spoken
    engine.runAndWait()   # Process the queued speech commands and wait until they are finished
    engine.stop()   # Stop the speech engine
```

The readn(nstr) function uses the pyttsx3 library to convert a given string (nstr) into speech. This function initializes a text-to-speech engine, sets its properties (like voice and rate), and then speaks the string.

pyttsx3.init() initializes a text-to-speech engine instance. This engine will be used to convert text to speech.
Setting the Voice Property

engine.setProperty('voice', "english+f5") sets the voice property of the engine.
"english+f5" is a specific identifier for a voice in the pyttsx3 library. Voices vary in accent, gender, and age, and different systems may have different voices installed.
You can list available voices on your system using engine.getProperty('voices') and choose the desired one.
Setting the Rate Property

engine.setProperty('rate', 130) sets the speech rate of the engine.
The rate is measured in words per minute (WPM). A rate of 130 WPM is relatively slow, making the speech clear and easy to understand.
Speaking the Text

engine.say(nstr) queues the text string nstr to be spoken by the engine.
The text is converted to speech when runAndWait() is called.
Running and Waiting

engine.runAndWait() processes all the currently queued commands, including speaking the text.
This function will block (wait) until all the speech commands are finished.
Stopping the Engine

engine.stop() stops the speech engine.
This is typically used to terminate the speech synthesis early, but in this case, it's called after runAndWait() to ensure that the engine is cleanly stopped after speaking.

## Additional Notes
### Voice Selection: The voice identifier "english+f5" may not be available on all systems. You can list available voices using engine.getProperty('voices') and select a suitable one.  

Error Handling: This function does not currently handle any exceptions. In a production environment, you may want to add error handling to manage cases where the speech engine fails to initialize or speak the text.

```python
## This fn is to calculate disease severity and term need of doctor or precautionary treatment : calculates disease severity based on symptoms experienced (exp) and the duration in days (days)
def calc_condition(exp,days):
    sum=0    # Accumulates the severity scores of the symptoms
# Implemented for loop to iterate n compute the severity of disease
# iterates over each symptom in the exp list. severityDictionary is a pre-defined dictionary that maps symptoms to their severity scores. For each symptom (item), its severity score is added to sum
    for item in exp:
         sum=sum+severityDictionary[item]
# sets how to proceed
    if((sum*days)/(len(exp)+1)>13):
        print("You should take the consultation from doctor. ")
    else:
        print("It might not be that bad but you should take precautions.")
```

```python
def getDescription():           # Define a function named getDescription with no parameters
    global description_list     # Declare 'description_list' as a global variable to access and modify it within the function
    with open('/content/symptom_Description.csv') as csv_file:
        # Open the CSV file located at '/content/symptom_Description.csv' in read mode
        # The 'with' statement ensures the file is properly closed after reading
        csv_reader = csv.reader(csv_file, delimiter=',')
        # Create a CSV reader object to read the file with ',' as the delimiter
        line_count = 0
        # Initialize a variable 'line_count' to keep track of the number of lines read from the file
        for row in csv_reader:
            # Iterate over each row in the CSV file using the CSV reader object
            _description = {row[0]: row[1]}
            # Create a dictionary '_description' with the first column as the key and the second column as the value
            description_list.update(_description)
            # Update the global 'description_list' dictionary with the '_description' dictionary
```

```python
def getSeverityDict():            # Define a function named getSeverityDict with no parameters
    global severityDictionary     # Declare 'severityDictionary' as a global variable to access and modify it within the function
    with open('/content/Symptom_severity.csv') as csv_file:
        # Open the CSV file located at '/content/Symptom_severity.csv' in read mode
        # The 'with' statement ensures the file is properly closed after reading
        csv_reader = csv.reader(csv_file, delimiter=',')        # Create a CSV reader object to read the file with ',' as the delimiter
        line_count = 0            # Initialize a variable 'line_count' to keep track of the number of lines read from the file
        try:
            for row in csv_reader:          # Try to iterate over each row in the CSV file using the CSV reader object
                _diction = {row[0]: int(row[1])}           # Create a dictionary '_diction' with the first column as the key and the second column (converted to an integer) as the value
                severityDictionary.update(_diction)         # Update the global 'severityDictionary' dictionary with the '_diction' dictionary
        except:
            pass           # If an exception occurs during the try block, silently pass without doing anything
```

```python
def getprecautionDict():                # Define a function named getprecautionDict with no parameters
    global precautionDictionary         # Declare 'precautionDictionary' as a global variable to access and modify it within the function
    with open('/content/symptom_precaution.csv') as csv_file:
        # Open the CSV file located at '/content/symptom_precaution.csv' in read mode
        # The 'with' statement ensures the file is properly closed after reading
        csv_reader = csv.reader(csv_file, delimiter=',')          # Create a CSV reader object to read the file with ',' as the delimiter
        line_count = 0               # Initialize a variable 'line_count' to keep track of the number of lines read from the file
        for row in csv_reader:       # Iterate over each row in the CSV file using the CSV reader object
            _prec = {row[0]: [row[1], row[2], row[3], row[4]]}      # Create a dictionary '_prec' with the first column as the key and a list of the next four columns as the value
            precautionDictionary.update(_prec)                      # Update the global 'precautionDictionary' dictionary with the '_prec' dictionary
```

```python
## Fn to input name and greet user
def getInfo():
    # Print a header for the HealthCare ChatBot for user to see
    print("-----------------------------------HealthCare ChatBot-----------------------------------")
    # Print a prompt asking for the user's name, with specific formatting
    print("\nYour Name? \t\t\t\t", end="->")
    # Wait for the user to enter their name and store the input in the variable 'name'
    name = input("")
    # Print a personalized greeting using the user's name
    print("Hello, ", name)

## Fn to take list of disease names (dis_list) n input string to search for within disease names (inp) n return list of predicted diseases (pred_list)
def check_pattern(dis_list, inp):
    # Initialize an empty list to store predicted diseases
    pred_list = []
    # Replaces spaces in inp with underscores to standardize the format
    inp = inp.replace(' ', '_')
    # Create a pattern string for the input (essentially the same as the modified input)
    patt = f"{inp}"
    # Compiles the regular expression pattern patt into a regexp object
    regexp = re.compile(patt)
    # Iterate through the disease list and add items to pred_list that match the pattern
    pred_list = [item for item in dis_list if regexp.search(item)]
    # Check if any matches were found : Uses a list comprehension to find all items in dis_list that match the regexp pattern and stores them in pred_list
    if len(pred_list) > 0:
        # Return 1 and the list of predicted diseases if matches were found
        return 1, pred_list
    else:
        # Return 0 and an empty list if no matches were found
        return 0, []
```

```python
## fn to classify symptoms into diseases using Decision Tree Classifier model
def sec_predict(symptoms_exp):                        # Define a function named sec_predict that takes a list of symptoms 'symptoms_exp' as an argument
    df = pd.read_csv('/content/Training.csv')         # Read the CSV file located at '/content/Training.csv' into a DataFrame 'df'
    X = df.iloc[:, :-1]                               # Select all columns except the last one as features 'X'
    y = df['prognosis']                               # Select the last column as the target variable 'y'
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
    # Split the data into training and testing sets with 70% for training and 30% for testing
    # The 'random_state' parameter ensures reproducibility

    rf_clf = DecisionTreeClassifier()                 # Initialize a Decision Tree Classifier 'rf_clf'
    rf_clf.fit(X_train, y_train)                      # Train the classifier on the training data 'X_train' and 'y_train'
    symptoms_dict = {symptom: index for index, symptom in enumerate(X)}
    # Create a dictionary 'symptoms_dict' that maps each symptom to an index
    # 'enumerate(X)' generates pairs of (index, symptom)
    input_vector = np.zeros(len(symptoms_dict))
    # Create an input vector of zeros with the same length as the number of symptoms

    for item in symptoms_exp:                          # Iterate over each symptom in the input list 'symptoms_exp'
      input_vector[[symptoms_dict[item]]] = 1        # Set the corresponding index in 'input_vector' to 1 based on 'symptoms_dict'
    return rf_clf.predict([input_vector])              # Use the trained classifier 'rf_clf' to predict the disease based on 'input_vector'     # Return the prediction result


## Fn to display predicted disease
def print_disease(node):
    # Define a function named print_disease that takes one argument 'node'
    node = node[0]
    # Assign the first element of 'node' to the variable 'node'.
    # This assumes 'node' is a list or array where the first element contains the data of interest.
    val = node.nonzero()
    # Find the indices of the non-zero elements in 'node' and assign them to 'val'.
    # 'nonzero()' returns a tuple of arrays, one for each dimension of 'node', containing the indices of the non-zero elements.
    disease = le.inverse_transform(val[0])
    # Use the label encoder 'le' to inverse transform the indices in 'val[0]' to their original string labels.
    # 'val[0]' accesses the first array in the tuple returned by 'nonzero()', which contains the indices of the non-zero elements.
    return list(map(lambda x: x.strip(), list(disease)))
    # Convert the 'disease' array to a list, strip any leading/trailing whitespace from each element using 'strip()',
    # and return the list. 'map()' applies the 'lambda' function to each element of 'disease', and 'list()' converts the result back to a list.
```

```python
## works sometimes but editable


# Define a function that translates a decision tree into code
# Loop initialised -> enter symptoms -> suggest feature matching -> choice chosen -> no. of days in -> severity check -> make 2 predictions to calculate disease -> description n precaution detection
def tree_to_code(tree, feature_names):
    tree_ = tree.tree_

    # Map feature indices to feature names
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    # Prepare the list of features (symptoms)
    chk_dis = ",".join(feature_names).split(",")
    symptoms_present = []

    # Start an infinite loop to get user input for symptoms
    while True:
        print("\nEnter the symptom you are experiencing  \t\t", end="->")
        disease_input = input("")

        # Check if the input symptom matches any feature name
        conf, cnf_dis = check_pattern(chk_dis, disease_input)

        # If there is a match
        if conf == 1:
            print("searches related to input: ")
            for num, it in enumerate(cnf_dis):
                print(num, ")", it)

            # Ask user to select the correct symptom from suggestions
            if num != 0:
                print(f"Select the one you meant (0 - {num}):  ", end="")
                conf_inp = int(input(""))
            else:
                conf_inp = 0

            disease_input = cnf_dis[conf_inp]
            break

        # If the input symptom is not valid, prompt again
        else:
            print("Enter valid symptom.")

    # Get the duration of the symptom from the user
    while True:
        try:
            num_days = int(input("Okay. From how many days ? : "))
            break
        except:
            print("Enter valid input.")

    # Define a recursive function to traverse the decision tree
    def recurse(node, depth):
        indent = "  " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]

            # Check if the current node matches the input symptom
            if name == disease_input:
                val = 1
            else:
                val = 0

            # Traverse left or right based on the threshold value
            if val <= threshold:
                recurse(tree_.children_left[node], depth + 1)
            else:
                symptoms_present.append(name)
                recurse(tree_.children_right[node], depth + 1)
        else:
            # When a leaf node is reached, determine the predicted disease
            present_disease = print_disease(tree_.value[node])
            print("You may have " , present_disease)

            # Get the list of symptoms for the predicted disease
            red_cols = reduced_data.columns
            symptoms_given = red_cols[reduced_data.loc[present_disease].values[0].nonzero()]
            dis_list = list(symptoms_present)

            # Display symptoms and ask the user for further confirmation
            if len(dis_list) != 0:
                print("symptoms present  " + str(list(symptoms_present)))
            print("symptoms given " + str(list(symptoms_given)))
            print("Are you experiencing any ")

            # Check which symptoms the user is experiencing
            symptoms_exp = []
            for syms in list(symptoms_given):
                inp = ""
                print(syms, "? : ", end='')
                while True:
                    inp = input("")
                    if(inp == "yes" or inp == "no"):
                        break
                    else:
                        print("provide proper answers i.e. (yes/no) : ", end="")
                if(inp == "yes"):
                    symptoms_exp.append(syms)

            # Make a second prediction based on user-confirmed symptoms
            second_prediction = sec_predict(symptoms_exp)
            print(second_prediction)

            # Calculate the condition severity
            calc_condition(symptoms_exp, num_days)

            # Display possible diseases and descriptions
            if present_disease[0] == second_prediction[0]:
                print("You may have ", present_disease[0])
                print(description_list[present_disease[0]])
                #readn(f"You may have",present_disease[0])
                #readn(description_list[present_disease[0]])
            else:
                print("You may have ", present_disease[0], "or ", second_prediction[0])
                print(description_list[present_disease[0]])
                print(description_list[second_prediction[0]])

            # Provide precautions for the predicted disease
            precution_list = precautionDictionary[present_disease[0]]
            print("Take following measures : ")
            for i, j in enumerate(precution_list):
                print(i + 1, ")", j)

            # Calculate and display the confidence level of the prediction
            confidence_level = (1.0 * len(symptoms_present)) / len(symptoms_given)
            print("confidence level is " + str(confidence_level))

    # Start the recursion from the root node
    recurse(0, 1)

# Fetch and prepare data required for the prediction
getSeverityDict()
getDescription()
getprecautionDict()
getInfo()

# Run the tree_to_code function with the classifier and column names
tree_to_code(clf, cols)
print("----------------------------------------------------------------------------------------")
```
